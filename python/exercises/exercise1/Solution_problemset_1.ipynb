```python
# Import the packages we need
import numpy as np
from matplotlib import pyplot as plt
from scipy import linalg as spla
import sys
sys.path.append(r"C:\Users\oiseth\Documents\GitHub\TKT4108StructuralDynamics2\python\modules")
from time_integration import * # Import all functions in the module time_integration

# Define stiffness amd mass matrix in nodal basis 
k = 10000
m = 1000
KK = np.array([[4, -2],[-2, 2]])*k
MM = np.array([[1, 0],[0, 1]])*m

# Solve generalized eigenvalue problem
lam,v = spla.eig(KK,MM)

# Sort eigenvalues and eigenvectors (lowest frequency first)
pos = np.argsort(lam)
lam = lam[pos]
v = v[:,pos]

# Normalize eigenvalues making y_1=1
v[:,0] = v[:,0] / v[0,0] 
v[:,1] = v[:,1] / v[0,1] 

# Transform the stiffness and mass matrices to modal basis
KK_modal = v.T @ KK @ v
MM_modal = v.T @ MM @ v

# Define modal damping matrix
zeta_1 = 0.5/100
zeta_2 = 0.5/100
lam = np.real(lam)
CC_modal = np.array([[2*MM_modal[0,0]*lam[0]**0.5*zeta_1, 0] ,[0, 2*MM_modal[1,1]*lam[1]**0.5*zeta_2]])

# Make time vector
h = 1/50 # time step
t = np.arange(0,10000)*h

# Generate the load
rho_X1_X2 = 0.5 # Load correlation coefficient
stdX1 = 1000.0 # Standard deviation X1
stdX2 = 1000.0 # Standard deviation X2

covmX = np.array(([[stdX1**2, rho_X1_X2*stdX1*stdX2], [rho_X1_X2*stdX1*stdX2, stdX2**2]])) # Covariance matrix of the loads
lam2,v2 = spla.eig(covmX) #Solve eigenvalue problem using scipy 
covmX_modal = np.matmul(np.matmul(v2.T,covmX),v2) # Transform covariance matrix to uncorrelated space

U = np.vstack((np.random.normal(0, covmX_modal[0,0]**0.5, t.shape[0]),np.random.normal(0, covmX_modal[1,1]**0.5, t.shape[0])))
X = np.matmul(v,U) # Transform to correlated space

# Transform load to modal basis
X_modal = v2.T @ X

u0 = np.zeros((2,1))
udot0 = np.zeros((2,1))

gamma = 1/2
beta = 1/4

z, zdot, z2dot = linear_newmark_krenk(MM_modal,CC_modal,KK_modal,X_modal,u0,udot0,h,gamma,beta)

cm = 1/2.54
fix, axs = plt.subplots(2,1,figsize=(15*cm,10*cm))
axs[0].plot(t,z[0,:])
axs[0].set_ylim(-0.2,0.2)
axs[0].grid(True)
axs[0].set_ylabel("$z_1$")

axs[1].plot(t,z[1,:])
axs[1].set_ylim(-0.2,0.2)
axs[1].grid(True)
axs[1].set_ylabel("$z_2$")
axs[1].set_xlabel("$t$")

plt.tight_layout()
```

### Problem 2: Technical terms
This problem will study some basic definitions used to characterize a stochastic process.

a) Explain the following terms and how they are related using text and figures
- Probability density function.
- Cumulative distribution function.
- Mean value
- Mean square value
- Variance

#### Solution:
The **probability density function** of a variable $X$ specify the probability of the variable beeing in a particular interval. The probability density is always positive and the area under the curve is always one to fulfill the law of total probability
$$ P\left[ a \leq X \leq b  \right] = \int_{a}^{b} f_X(x) dx $$

The **cumulative distribution function** is the integral of the probability density function. The function is zero for $X=-\infty$ and 1 for $X=\infty$
$$F_X(x) = P\left[ X \leq x  \right] = \int_{-\infty}^{x}f_X(x)dx $$

The **mean value** is the centroid of the probability density function (first moment) or the average

$$ \mu_x = \int_{-\infty}^{\infty}xf_X(x)dx $$

$$ \mu_x = \frac{1}{T}\int_{0}^{T}x(t)dt $$

The **mean square value** is the second moment of the probability density function or the time average of the square of the stochastic process

$$ E\left[x^2 \right] = \int_{-\infty}^{\infty}x^2f_X(x)dx $$

$$ E\left[x^2 \right] = \frac{1}{T}\int_{0}^{T}x^2(t)dt $$

The **variance** is the same as the mean square, but with the mean value removed

$$ \sigma_x^2 = \int_{-\infty}^{\infty} \left( x-\mu_x \right)^2 f_X(x)dx $$

$$ \sigma_x^2 = \frac{1}{T}\int_{0}^{T} \left(x(t) -\mu_x \right)^2dt $$





b) Study Figure 2.3 in Newland and explain the conditional probability
Which requirements need to be fulfilled for x and y to be independent?

$$ p(x|y) = \frac{p(x,y)}{p(y)} $$

#### Solution
The figure below shows the joint probability distribution of $x_1$ and $x_2$. The volume under the surface is equal to one. 

<img src="Conditional_probability.jpg">

We now consider that we know the outcome of $x_2$, this means that the probability (elementary volume) $p(x_1,x_2)dx_2dx_2$ needs to be within the shaded volume. We also need to "update the reference volume" since the volume of the strip shall be the reference instaed of the volume under the entire joint probability function. This can be expressed as follows. 

$$ p({x_1}|{x_2})d{x_1} = \frac{\text{Elementary volume}}{\text{Volume of shaded strip}} $$

Inserting the metamatical expressions provides

$$ p({x_1}|{x_2})d{x_1} = \frac{{p({x_1},{x_2})d{x_1}d{x_2}}}{{d{x_2}\int_{ - \infty }^\infty  {p({x_1},{x_2})d{x_1}} }} $$

Which can be further simplified to

$$p({x_1}|{x_2}) = \frac{{p({x_1},{x_2})}}{{\int_{ - \infty }^\infty  {p({x_1},{x_2})d{x_1}} }}$$

integrating $\int_{ - \infty }^\infty  {p({x_1},{x_2})d{x_1}}$ results in $p({x_2})$ providing the following expression for conditional probability

$$p({x_1}|{x_2}) = \frac{{p({x_1},{x_2})}}{{p({x_2})}}$$

The variables $x_1$ and $x_2$ are independent if the joint probability distribution can be expressed as the product of the marginal distributions $p(x,y)=p(x) p(y)$. If the outcome for one random variable is known, it does not influence the conditional probability of the other variable, thus $p(x|y)=p(x)$.
  
c) Explain the following two terms and how they are related.
- Autocorrelation
- Cross-correlation

#### Solution
Let us consider the two variables $x_1$ and $x_2$
The auto-correlation functions of for instance the stochastic process $x_1$ is the expected value of the function at one time instant multiplied with itself at another time instant

$$R_{x_1}(t_1,t_2)= E \left[x_1(t_1)x_1(t_2)  \right] $$

If the stochastic prosess is stationary, the auto-correlation is only dependen on the time lag $\tau=t_2-t_1$. The auto.correlation function can then be written as

$$R_{x_1}(\tau)= E \left[x_1(t)x_1(t+\tau)  \right] $$

The auto-correlation is in most cases a decaying function that is symmetic about the vertical axis

The cross-correlation function is very similar to the auto-correlation function, but in this case it is the product of two different stochastic processes. 

$$R_{x_1x_2}(\tau)= E \left[x_1(t)x_2(t+\tau)  \right] $$

The cross-correlation function is in general not symmetric about the vertical axis.


```python
tau = np.linspace(-10,10,1000)
mean_x1 = 0.5
mean_x2 = 0.7
Rx1 = np.sin(np.pi*tau)/tau + mean_x1**2 

Rx1x2 = np.sin(np.pi*(tau+1))/(tau+1) + mean_x1*mean_x2

plt.figure()
plt.plot(tau,Rx1)
plt.title("Typical auto-correlation function")
plt.ylabel("$R_{x_1}(\tau)$")
plt.grid()

plt.figure()
plt.plot(tau,Rx1x2)
plt.title("Typical cross-correlation function")
plt.ylabel("$R_{x_1x_2}(\tau)$")
plt.grid()

```

### Problem 3: Rayleigh distribution

The cumulative Rayleigh distribution is defined as follows.

$$F_x(x,\sigma) = 1-e^{-x^{2}/(2\sigma ^{2})} $$

a) Obtain the probability density function $f_x(x)$ by taking the derivative of the cumulative distribution. Confirm that the area under the curve is unity by numerical integration in python and explain why this is essential. Assume $\sigma = 2$ when performing the numerical integration

#### Solution
Taking the derivative of the cumulative distribution to obtain the probability density function

$$ f_x(x) =  F^\prime_x(x)=  e^{-x^{2}/(2\sigma ^{2})} x/\sigma ^{2}$$

The area under the curve needs to be one to fulfill the total probability theorem.

Numerical intagration in python to confirm that the derived expression is correct.

```python
x = np.linspace(0,10,1000)

sigma = 2

fx = np.exp(-x**2/2/sigma/2)*x/sigma**2

plt.figure()
plt.plot(x,fx)
plt.title("Rayleigh probability density function")
plt.grid()
plt.xlabel("$x$")
plt.ylabel("$f_x(x)$")

print("Total probability = " + str(np.trapz(fx,x)))
```
