# Example x-1. The Nyquist shanning sampling theorem

```python
import numpy as np
from matplotlib import pyplot as plt
tc = np.arange(0,100,1/10000)

f = 10
fs = 15
yc = np.sin(2*np.pi*tc*f-np.pi/10)

t = np.arange(0,1,1/fs)
y = np.sin(2*np.pi*t*f-np.pi/10)

fig, axs = plt.subplots(1,2)

yfft = np.fft.fft(y)

axs[0].plot(tc,yc)
axs[0].set_xlim(0,1)
axs[0].plot(t,y,'-o')
```
