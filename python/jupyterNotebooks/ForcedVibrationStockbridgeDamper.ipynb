```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e = 0.01
L = 0.5
m = 20.0

t = np.array([np.linspace(0,0.1,1000)])
omega_n = 20*2*np.pi
u = e*np.cos(omega_n*t) # Displacement

h = e*np.sin(omega_n*t) # Height of triangle
alpha = np.arcsin(h/L) # Angle at mass
LT = L*np.sin(alpha)+u # Length of horizontal part of the triangle 
em = LT*np.sin(alpha) # Moment arm 
# Forces
fi = -omega_n**2*m*e*np.cos(omega_n*t) # Inertia force
S = fi/np.cos(alpha) #Force in rod
M = S*em # Moment

plt.figure()
plt.plot(t[0,:], u[0,:]);
plt.plot(t[0,:], h[0,:]);
plt.plot(t[0,:], em[0,:]);
plt.grid()


plt.figure()
plt.plot(t[0,:], fi[0,:]);
plt.plot(t[0,:], S[0,:]);
plt.grid()


plt.figure()
plt.plot(t[0,:], M[0,:]);
plt.grid()



```
