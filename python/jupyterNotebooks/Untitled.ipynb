# Note book example

This is an example of how to make a notebook $\alpha + \beta =1$

$$ x^2+x+2=0 $$



```python
import numpy as np
from matplotlib import pyplot as plt

t = np.linspace(0.0,10.0,1000)

f = 0.2

x = np.sin(np.pi*f*t)


plt.figure()
plt.plot(t,x)
plt.grid()
plt.xlabel(r'$t$')
plt.ylabel(r'$x(t)$')
```
